**Q02. Given the following data, create a line chart with the following properties:**<br>
      
      1. Chart title:'Weekly Temperature'; label for X axis:'Week Day'; label for Y axis:'Tempearue in Fahrenheit'
      2. Add line color, line width, line style, marker, and alpha of your choice
      3. Add legend and grid
      4. Use output image as reference
**Output image:** <br>      


```python
week_day = ['SUN','MON','TUE','WED','THU','FRI','SAT']
high_temp = [88, 97, 95, 98, 100, 101, 93]
low_temp = [77, 80, 79, 73, 80, 81, 78]
avg_temp = [82, 88, 87, 85, 90, 91, 86]
plt.plot(week_day,high_temp,label = 'High', color = 'blue', linestyle='-.', marker= 'o')
plt.plot(week_day,low_temp, label = 'Low', color = 'red', linestyle='-.', marker= 'o')
plt.plot(week_day,avg_temp, label = 'Avg', color = 'green', linestyle='-.', marker= 'o')
plt.title('Weekly Temperature')
plt.xlabel('Day of the week')
plt.ylabel('Temperature in Fahrenheit')
plt.grid(True)
plt.legend(loc = 'lower right')
plt.show
```

**Q03. Use the data below, create a stacked bar chart. Use the output image as reference**<br>
years = range(2015,2020) <br>
class_counts = {
    'Face to Face':np.array([47, 50, 48, 40, 35]),
    'Online': np.array([29, 28, 40, 45, 50]),
}<br><br>
**Output image:** <br>


```python
years = range(2015,2020)
class_counts = { 'Face to Face':np.array([47, 50, 48, 40, 35]), 'Online': np.array([29, 28, 40, 45, 50])}
bar1 = plt.bar(years, class_counts['Face to Face'], label = 'Face to Face')
bar2 = plt.bar(years, class_counts['Online'], bottom = class_counts['Face to Face'], label = 'Online')
plt.bar_label(bar1, label_type = 'center')
plt.bar_label(bar2, label_type = 'center')
plt.xlabel('Years')
plt.ylabel('Number of Classes')
plt.title('Class Modality')
plt.grid()
plt.legend()
plt.tight_layout()
plt.show
```

**Q04. Load dataset 'iris' from Seaborn and create a stacked Histogram chart to see the distribution of sepal length. Use the output image as reference.** <br><br>
**Output image:**<br>


```python
import seaborn as sns
flowers_df = sns.load_dataset("iris")
print(flowers_df)
plt.figure(figsize=(8,6))
palette = {
    "setosa":    "tab:blue",   # normally green slot (was virginica)
    "versicolor":"tab:orange",
    "virginica": "tab:green"     # swap into the blue slot
}
sns.histplot(data= flowers_df, x = 'sepal_length', hue = 'species', hue_order = ['virginica', 'versicolor', 'setosa'], palette = palette, binrange = (4.0, 6.8), binwidth = 0.24,alpha=0.95, multiple = 'stack', edgecolor = None)
plt.title('Distribution of Sepal Length')
plt.xlabel('')
plt.ylabel('')
plt.ylim(0, 24)
plt.grid()
plt.show()
```

**Q05. Load dataset 'tips' from Seaborn and create the bar chart (use Seaborn barplot() method) to display bill amount on each day based on gender of servers. Use the output image as reference.**<br><br>
**Output image:** <br>


```python
tips_df = sns.load_dataset('tips')
print(tips_df)
sns.barplot(x='day', y='total_bill', data = tips_df, hue='sex', 
                palette = ['blue','red'])
plt.grid(c='gray', linestyle= 'dotted'),
plt.xlabel('day')
plt.ylabel('Total Bill')
plt.show
```

**Q07. Load tips dataset and create scatter chart(Seaborn scatterplot() method) to see the relationship between tip and bill amount using day as hue. Use the output image as reference.**<br><br>
**Output image:** <br>


```python
tips_df = sns.load_dataset('tips')
print(tips_df)
sns.scatterplot(x='tip', y = 'total_bill', data=tips_df, hue='day', 
                palette ='deep')
plt.grid(c='gray', linestyle= 'dotted')
plt.title('Tip vs. Total bill')
plt.xlabel('tip')
plt.ylabel('total_bill')
plt.show
```

**Q08. Create a pie chart based on the data below to show class modality and its weight. Use the output image as reference.** <br>
modality = ['Face to Face', 'Online Anytime', 'Online with Schedule'] <br>
weight = [30, 50, 20]<br><br>
**Output image:**<br>


```python
modality = ['Face to Face', 'Online Anytime', 'Online with Schedule']
weight = [30, 50, 20]
sns.set_style('darkgrid')
explode = (0.1,0,0)
pie = plt.pie(weight, explode=explode, labels=modality, autopct = '%1.1f%%', colors=['yellow','orange','cyan'], shadow=True, startangle= 140,  wedgeprops={         # turn off edges
        "edgecolor": "none",
        "linewidth": 0
    });
plt.title('Class Modality for Fall 2024')
```
